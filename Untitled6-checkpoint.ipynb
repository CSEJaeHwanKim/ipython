```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

####>>> s = MySet([1, 2, 3]) 
####>>> t = MySet([3, 4, 5]) 
####>>> s = s - t 
####>>> print s 
####MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    def union(self, A):
        for x in A:
            if x not in self:
                self.append(x)
        return MySet(self)
    def intersection(self, B):
        w=[]
        for x in self:
            if x in B and x not in w:
                w.append(x)
        return MySet(w)
    def difference(self, C):
        w=[]
        for x in self:
            if x not in C:  
                w.append(x)
        return MySet(w)
    def __or__(self,D):
        return self.union(D)
    def __sub__(self,E):
        return self.difference(E)
    def __and__(self,F):
        return self.intersection(F)
    

s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
w1=s|t
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
w2=s&t
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
w3=s-t
print w1
print w2
print w3
```
